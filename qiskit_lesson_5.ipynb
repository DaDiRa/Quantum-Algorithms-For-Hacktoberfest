```python
from qiskit import *
```

```python
circuit = QuantumCircuit(3,3)
```

```python
%matplotlib inline
circuit.draw(output='mpl')
```

```python
#setting the value of the qubit I will teleport at |1>
circuit.x(0)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
#qunatum entanglement
circuit.h(1)
circuit.cx(1,2)
```

```python
circuit.draw(output = 'mpl')
```

```python
#So basically now I move q2 very far away from q0 and q1, to a place where the speed of light takes minutes to reach
#and I continue to apply the quantum teleportation protocol on each location
circuit.cx(0,1)
circuit.h(0)
circuit.draw(output = 'mpl')
```

```python
#measure the qubits at one location
circuit.barrier()
circuit.measure([0,1],[0,1])
circuit.draw(output='mpl')
```

```python
#So called "G" operation on the second location. This operator can change according to the measurement above.
#Because you need to know what one location measured in order to apply the correct G operator on the other location
#the light speed limit of information exchange is guaranteed.
circuit.barrier()
circuit.cx(1,2)
circuit.cz(0,2)
circuit.draw(output='mpl')
```

```python
#simulate and measure, notice that c2 is always 1
circuit.measure(2,2)
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = simulator, shots = 1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
